```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("data_windows")

# Load small sample of authentication logs
```

```python
import pandas as pd

redteam_df = pd.read_csv(
   DATA_PATH / "redteam.txt.gz",
    compression="gzip",
    header=None
)

redteam_df.columns = [
    "timestamp",
    "user",
    "source_computer",
    "destination_computer"
]

redteam_df["timestamp"] = redteam_df["timestamp"].astype(int)

rt_min = redteam_df["timestamp"].min()
rt_max = redteam_df["timestamp"].max()

print("Redteam min:", rt_min)
print("Redteam max:", rt_max)
```

```python
BUFFER = 10000
start_time = rt_min - BUFFER
end_time = rt_max + BUFFER
```

```python
import pandas as pd

max_normal = 500_000     # limit normal samples
max_attack = 200_000     # limit attack samples

normal_rows = []
attack_rows = []

normal_count = 0
attack_count = 0

for chunk in pd.read_csv(
    DATA_PATH / "auth.txt.gz",
    compression="gzip",
    header=None,
    usecols=[0,1,3,4,8],
    chunksize=1_000_000
):

    chunk.columns = [
        "timestamp",
        "source_user",
        "source_computer",
        "destination_computer",
        "success"
    ]

    chunk["timestamp"] = chunk["timestamp"].astype(int)

    # Merge with redteam by matching timestamp + src + dst
    merged = chunk.merge(
        redteam_df,
        how="left",
        on=["timestamp", "source_computer", "destination_computer"]
    )

    chunk["label"] = merged["user"].notna().astype(int)

    attack_part = chunk[chunk["label"] == 1]
    normal_part = chunk[chunk["label"] == 0]

    if attack_count < max_attack:
        needed = max_attack - attack_count
        attack_rows.append(attack_part.head(needed))
        attack_count += len(attack_part.head(needed))

    if normal_count < max_normal:
        needed = max_normal - normal_count
        normal_rows.append(normal_part.head(needed))
        normal_count += len(normal_part.head(needed))

    if attack_count >= max_attack and normal_count >= max_normal:
        break

train_df = pd.concat(attack_rows + normal_rows, ignore_index=True)

print("Final dataset size:", len(train_df))
print("Attack samples:", train_df["label"].sum())
```

```python
auth_chunks = []

for chunk in pd.read_csv(
    DATA_PATH / "auth.txt.gz",
    compression="gzip",
    header=None,
    usecols=[0,1,3,4,8],  # only needed columns
    chunksize=1_000_000
):
    
    chunk.columns = [
        "timestamp",
        "source_user",
        "source_computer",
        "destination_computer",
        "success"
    ]
    
    chunk["timestamp"] = chunk["timestamp"].astype(int)
    
    filtered = chunk[
        (chunk["timestamp"] >= start_time) &
        (chunk["timestamp"] <= end_time)
    ]
    
    auth_chunks.append(filtered)

auth_window = pd.concat(auth_chunks, ignore_index=True)

print("Filtered auth rows:", len(auth_window))
```

```python
auth_window["user"] = auth_window["source_user"].str.split("@").str[0]
auth_window["success"] = auth_window["success"].map({"Success":1, "Failure":0})

auth_window = auth_window[[
    "timestamp",
    "user",
    "source_computer",
    "destination_computer",
    "success"
]]

auth_window["event_type"] = "auth"
auth_window["event_id"] = auth_window.index
```

```python
flow_chunks = []

for chunk in pd.read_csv(
    DATA_PATH / "flows.txt.gz",
    compression="gzip",
    header=None,
    usecols=[0,1,2,4,7,8],  # only needed columns
    chunksize=1_000_000
):
    
    chunk.columns = [
        "timestamp",
        "duration",
        "source_computer",
        "destination_computer",
        "packet_count",
        "byte_count"
    ]
    
    chunk["timestamp"] = chunk["timestamp"].astype(int)
    
    filtered = chunk[
        (chunk["timestamp"] >= start_time) &
        (chunk["timestamp"] <= end_time)
    ]
    
    flow_chunks.append(filtered)

flow_window = pd.concat(flow_chunks, ignore_index=True)

flow_window.rename(columns={
    "packet_count": "packets",
    "byte_count": "bytes"
}, inplace=True)

flow_window["event_type"] = "flow"
flow_window["event_id"] = flow_window.index

print("Filtered flow rows:", len(flow_window))
```

```python
import os
os.makedirs("data_window/processed", exist_ok=True)

auth_window.to_parquet("data_windows/processed/auth_window.parquet")
flow_window.to_parquet("data_windows/processed/flow_window.parquet")
redteam_df.to_parquet("data_windows/processed/redteam.parquet")
```
